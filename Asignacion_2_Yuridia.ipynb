```python
import pandas as pd
data = pd.read_csv("SoftwareProject_DataSet.csv")
data.head()
```

```python
###Declaramos una variable para sacar el tamaño de la  poblacion.
```

```python
population_size =len(data)
print("El tamaño de la poblacion es", population_size)
```

```python
###Tomaremos una muestra aleatoria de un conjunto de datos.
```

```python
sample_size = 100
sample = data.sample(n = sample_size, random_state= 81)
```

```python
###Mostrar el dataFrame y ver el rango de los datos
```

```python
display(sample)
```

```python
###Verificamos si la muestra es correcta
```

```python
##Mi analisis sera sobre la comparativa de costos entre diferentes tipos de programacion
##para ello primero iniciaremos con un muestreo aleatorio sistematico.
```

```python
sample_size = 150
interval_k1 = population_size//sample_size
```

```python
systematic_sample= data.iloc[::interval_k1]
```
